```python
#step 3
data["Date"] = pd.to_datetime(data[["Yr","Mo","Dy"]].astype(str).agg('-'.join, axis=1))
data = data.drop(columns=["Yr","Mo","Dy"])
data.head()
```

```python
#step 4
data["Date"] = np.where(pd.DatetimeIndex(data["Date"]).year < 2000,data.Date,data.Date - pd.offsets.DateOffset(years=100))
data.head()
```

```python
#step 5
data = data.set_index("Date")
data.index.astype("datetime64[ns]")
```

```python
#step 6
data.isnull().values.ravel().sum()
```

```python
#step 7
data.count().sum()
```

```python
#step 8
pd.DataFrame(data.mean())
```

```python
#step 9
def stats(stats):
    stats = pd.Series(stats)
    Min = stats.min()
    Max = stats.max()
    Mean = stats.mean()
    Std = stats.std()
    res = [Min,Max,Mean,Std]
    indx = ["Min","Max","Mean","Std"]
    res = pd.Series(res,index=indx)
    return res
loc_stats = data.apply(stats)
pd.DataFrame(loc_stats)
```

```python
#step 10
day_stats = data.apply(stats,axis=1)
pd.DataFrame(day_stats.head())
```

```python
#step 11
jan_data = data[data.index.month == 1]
pd.DataFrame(jan_data.mean())
```

# Question 6 - Lineplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
USMarandDiv = pd.read_csv("C:/Users/USER/us-marriages-divorces-1867-2014.csv")
USMarandDiv.head()
```

```python
years = USMarandDiv['Year']
marriages = USMarandDiv['Marriages_per_1000']
divorces = USMarandDiv['Divorces_per_1000']
USMarandDiv = plt.figure(figsize=(18,9))
USMarandDiv = plt.plot(years, marriages, label='Marriages')
USMarandDiv = plt.plot(years, divorces, label='Divorces')
USMarandDiv = plt.title("Number of marriages and divorces per capita in the U.S. between 1867 and 2014")
USMarandDiv = plt.xlabel("Years",fontsize=16)
USMarandDiv = plt.legend(fontsize = 14, loc = "upper left")
USMarandDiv = plt.ylabel("Marriages",fontsize=15)
USMarandDiv = plt.grid(True)
USMarandDiv
```

# Question 9 - Pie Chart 

```python
import pandas as pd
import matplotlib.pyplot as mplt
roman_emperor = pd.read_csv('roman-emperor-reigns.csv')
assemperors = roman_emperor[
    roman_emperor['Cause_of_Death'].apply(lambda x: 'assassinated' in x.lower())]

numassassinated = len(assemperors)
otherdeath = len(roman_emperor) - numassassinated
print(assemperors)
print(numassassinated)
print(otherdeath)
emperor = assemperors["Emperor"]
Cause_of_Death = assemperors["Cause_of_Death"]
mplt.pie(range(len(Cause_of_Death)), labels=emperor,autopct='%1.1f%%', startangle=80, radius=0.08 * 100)


mplt.show()
```

# Question 10 - Scatter Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
AddressRead = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv')
YearGroup = AddressRead.groupby('Year')

for name, group in YearGroup:
    plt.plot(group['Total Arcade Revenue (billions)'], group['Computer Science Doctorates Awarded (US)'], marker='o', linestyle='', markersize=12, label=name)

plt.legend()
plt.xlabel("Revenue")
plt.ylabel("Awards")
plt.show()
```
